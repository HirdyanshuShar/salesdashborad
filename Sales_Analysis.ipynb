```python
clean_df = [
    'Amount',
    'AmountIncludingVAT',
    'Description',
    'DocumentNo',
    'BilltoCustomerNo',
    'Quantity',
    'PostingDate',
    'No',
    'LocationCode',
    'GSTAssessableValueLCY',
    'InvDiscountAmount',
    'ItemCategoryCode',

]

df_small = df[clean_df]
```

```python
# =====================================================
# CLEAN NUMERIC COLUMNS
# =====================================================

numeric_cols = [

    'Amount',
    'AmountIncludingVAT',
    'Quantity',
    'GSTAssessableValueLCY',
    'InvDiscountAmount'
]

for col in numeric_cols:

    df_small[col] = (

        df_small[col]
        .astype(str)

        .str.replace(',', '', regex=False)

        .str.replace('₹', '', regex=False)

        .str.strip()
    )

    df_small[col] = pd.to_numeric(
        df_small[col],
        errors='coerce'
    )
```

```python
customer_df = df_small.groupby(
    'BilltoCustomerNo'
).agg({

    # NUMERIC COLUMNS
    'Amount': 'sum',
    'AmountIncludingVAT': 'sum',
    'Quantity': 'sum',
    'GSTAssessableValueLCY': 'sum',
    'InvDiscountAmount': 'sum',

    # TEXT COLUMNS
    'Description': lambda x:
        ', '.join(
            x.dropna()
            .astype(str)
            .unique()
        ),

    'DocumentNo': lambda x:
        ', '.join(
            x.dropna()
            .astype(str)
            .unique()
        ),

    'No': lambda x:
        ', '.join(
            x.dropna()
            .astype(str)
            .unique()
        ),

    'ItemCategoryCode': lambda x:
        ', '.join(
            x.dropna()
            .astype(str)
            .unique()
        ),

    # OTHER INFO
    'LocationCode': 'first',

    'PostingDate': 'max'

}).reset_index()
```

```python
import plotly.graph_objects as go
import pandas as pd

# ---------- Formats ----------

# Crore Format with ₹
def to_crore(x):
    return f"₹ {x/1e7:.2f} Cr"

# Indian Comma Format with ₹
def indian_format(x):
    x = int(round(x))
    s = str(x)

    if len(s) <= 3:
        return "₹ " + s

    last3 = s[-3:]
    rest = s[:-3]

    parts = []

    while len(rest) > 2:
        parts.insert(0, rest[-2:])
        rest = rest[:-2]

    if rest:
        parts.insert(0, rest)

    return "₹ " + ",".join(parts) + "," + last3


# ---------- Data ----------
df = top_products.copy()

# New Columns
df['Amount_Cr'] = df['Amount'].apply(to_crore)
df['Rounded_Amount'] = df['Amount'].apply(indian_format)

# ---------- TOTAL ROW ----------
total_amount = df['Amount'].sum()

total_row = pd.DataFrame({
    'ItemCategoryCode': ['TOTAL'],
    'Amount_Cr': [to_crore(total_amount)],
    'Rounded_Amount': [indian_format(total_amount)]
})

# ---------- Combine ----------
final_df = pd.concat([df, total_row], ignore_index=True)

# ---------- Row Styling ----------

# Light Green Rows (Logo Theme)
row_colors = ["#DDF5E3"] * len(final_df)

# TOTAL Row Dark Green
row_colors[-1] = "#008037"

# ---------- Bold TOTAL Row ----------
item_col = list(final_df['ItemCategoryCode'])
amount_cr_col = list(final_df['Amount_Cr'])
rounded_col = list(final_df['Rounded_Amount'])

item_col[-1] = f"<b>{item_col[-1]}</b>"
amount_cr_col[-1] = f"<b>{amount_cr_col[-1]}</b>"
rounded_col[-1] = f"<b>{rounded_col[-1]}</b>"

# ---------- Table ----------
fig = go.Figure(data=[go.Table(

    columnwidth=[280, 180, 220],

    header=dict(
        values=[
            '<b>Item Category</b>',
            '<b>Amount</b>',
            '<b> Full Rounded Amount</b>'
        ],

        # OSWAL Logo Inspired Green
        fill_color="#008037",

        font=dict(
            color='white',
            size=14
        ),

        align=['center'],
        height=38
    ),

    cells=dict(

        values=[
            item_col,
            amount_cr_col,
            rounded_col
        ],

        fill_color=[
            row_colors,
            row_colors,
            row_colors
        ],

        # align=['left','right','right'],
        align = ['center'],

        font=dict(
            color=[
                ['black']*(len(final_df)-1) + ['white'],
                ['black']*(len(final_df)-1) + ['white'],
                ['black']*(len(final_df)-1) + ['white']
            ],
            size=12
        ),

        height=34
    )
)])

# ---------- Layout ----------
fig.update_layout(

    title={
        'text': '<b>Top Products with Total Sales Amount</b>',
        'x': 0.5,
        'xanchor': 'center',
        'font': dict(
            size=22,
            color="#0E0E0E"
        )
    },

    paper_bgcolor='white',
    plot_bgcolor='white',

    width=950,
    height=550,

    margin=dict(
        l=20,
        r=20,
        t=70,
        b=20
    )
)

fig.show()
```

. TOP 10 PRODUCTS by QUANTITY SOLD

```python
import pandas as pd
import plotly.express as px

# ---------- Required Columns ----------
temp_df = df_small[
    ['ItemCategoryCode', 'Quantity', 'Amount']
].copy()

# ---------- Remove Missing Values ----------
temp_df = temp_df.dropna(
    subset=['ItemCategoryCode', 'Quantity', 'Amount']
)

# ---------- Convert Numeric ----------
temp_df['Quantity'] = pd.to_numeric(
    temp_df['Quantity'],
    errors='coerce'
)

temp_df['Amount'] = pd.to_numeric(
    temp_df['Amount'],
    errors='coerce'
)

# ---------- Group Data ----------
top_products = (
    temp_df.groupby(
        'ItemCategoryCode',
        observed=True
    )
    .agg({
        'Quantity': 'sum',
        'Amount': 'sum'
    })
    .reset_index()
)

# ---------- Top 10 Categories ----------
top_products = top_products.nlargest(
    10,
    'Quantity'
)

# ---------- Convert Quantity to Lakhs ----------
top_products['Quantity_Lakh'] = (
    top_products['Quantity'] / 100000
)

# ---------- Contribution % ----------
total_amount = top_products['Amount'].sum()

top_products['Contribution_%'] = (
    top_products['Amount'] / total_amount
) * 100

# ---------- Sort ----------
top_products = top_products.sort_values(
    by='Quantity_Lakh',
    ascending=False
)

# ---------- Colors ----------
dark_green = '#008037'
cream = '#F8F5E9'
black = '#1C1C1C'
blue = '#1565C0'

# ---------- Max Y Axis ----------
max_val = top_products['Quantity_Lakh'].max()

# ---------- Bar Chart ----------
fig = px.bar(

    top_products,

    x='ItemCategoryCode',
    y='Quantity_Lakh',

    text='Quantity_Lakh',

    title='Top 10 Product Categories by Quantity Sold'
)

# ---------- Styling ----------
fig.update_traces(

    marker=dict(
        color=dark_green,
        line=dict(
            width=1.5,
            color='white'
        )
    ),

    texttemplate='%{text:.2f} L',
    textposition='outside',

    # ---------- BLUE VALUES ABOVE BAR ----------
    textfont=dict(
        color=blue,
        size=13
    ),

    customdata=top_products[
        ['Quantity', 'Amount', 'Contribution_%']
    ],

    hovertemplate=
    '<b>Category:</b> %{x}<br>' +
    '<b>Full Quantity:</b> %{customdata[0]:,.0f}<br>' +
    '<b>Total Amount:</b> ₹ %{customdata[1]:,.0f}<br>' +
    '<b>Contribution:</b> %{customdata[2]:.2f}%<br>' +
    '<extra></extra>'
)

# ---------- Layout ----------
fig.update_layout(

    paper_bgcolor=cream,
    plot_bgcolor=cream,

    title=dict(
        x=0.5,
        font=dict(
            size=26,
            color=dark_green,
            family='Arial Black'
        )
    ),

    # ---------- X Axis ----------
    xaxis=dict(
        title='Product Category',
        tickangle=-20,

        tickfont=dict(
            size=12,
            color=black
        )
    ),

    # ---------- Y Axis ----------
    yaxis=dict(

        title='Quantity Sold (Lakhs)',

        tickmode='linear',

        dtick=6,          # gap of 6 Lakhs
        tick0=0,

        range=[0, max_val + 6],

        ticksuffix=' L',

        showticklabels=True,
        showgrid=True,

        gridcolor='rgba(0,0,0,0.10)',

        tickfont=dict(
            size=12,
            color=black
        )
    ),

    font=dict(
        family='Arial',
        color=black
    ),

    height=650,
    width=1150,

    margin=dict(
        l=90,
        r=40,
        t=90,
        b=90
    ),

    hoverlabel=dict(
        font_size=13
    )
)

# ---------- Show ----------
fig.show()
```

✅ Customer Contribution % (Donut Chart)

```python
import pandas as pd
import plotly.express as px

# ---------- Customer Wise Sales ----------
customer_share = (

    df_small.groupby('BilltoCustomerNo')['Amount']
    .sum()
    .reset_index()

)

# ---------- Top 5 + Others ----------
df = customer_share.copy()

top5 = df.nlargest(5, 'Amount').copy()

others = pd.DataFrame({

    'BilltoCustomerNo': ['Others'],

    'Amount': [
        df['Amount'].sum() - top5['Amount'].sum()
    ]
})

df_clean = pd.concat(
    [top5, others],
    ignore_index=True
)

# ---------- Convert Revenue ----------
df_clean['Amount_Cr'] = (
    df_clean['Amount'] / 10000000
)

# ---------- Contribution % ----------
df_clean['Contribution_%'] = (

    df_clean['Amount']
    / df_clean['Amount'].sum()

) * 100

# ---------- Colors ----------
colors = px.colors.qualitative.Pastel

# ---------- Donut Chart ----------
fig = px.pie(

    df_clean,

    names='BilltoCustomerNo',

    values='Contribution_%',

    hole=0.6,

    color_discrete_sequence=colors,

    title='Customer Contribution Analysis'
)

# ---------- Styling ----------
fig.update_traces(

    textinfo='percent',

    textposition='inside',

    customdata=df_clean[['Amount']],

    hovertemplate=

    "<b>Customer:</b> %{label}<br>" +
    "<b>Contribution:</b> %{value:.2f}%<br>" +
    "<b>Revenue:</b> ₹ %{customdata[0]:,.0f}<br>" +
    "<extra></extra>"
)

# ---------- Layout ----------
fig.update_layout(

    showlegend=True,

    height=650,
    width=1000,

    paper_bgcolor='white',

    title=dict(
        x=0.5,
        font=dict(
            size=24
        )
    )
)

# ---------- Show ----------
fig.show()
```
