Object Oriented way

Matplotlib allows the aspect ratio,DPI and figure size to be specified when the figure object is created. You can use fig size is a tuple of width and height of figure in inches 
dpi is the dots per inch (pixel per inch).

figsize is a tuple of width and height of figure in inches dpi is dots per-inch(pixel per inch).

```python
fig = plt.figure(figsize=(20,8), dpi=100)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

Boxplot and violinplot

boxplot and violinplot boxplots and violinplots are used to shown the distribution of quantiative data in a way that facilitates comparosons between variables or across levels of  a categorical variable. The box shows the quartiles of the dataset while whiskers extend to show the rest of distribution,except for points that are determined to be "outliers" using a method that is a function of inter-quartile range.

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

stripplot and swarmplot

the stripplot will draw a scatterplot where one variable is categorical.A strip plot can be drawn on its own,but it is also a good complemenr to a box or violin plot  in cases you want to show all observations along with some representation of underlying distribution

the swarm plot is similar to striplot(), but points are adjusted so they donot overlap.

the swarmplot is similar to stripplot(), so the points are adjusted dont overlap.This  gives a better representatin of distribution of values,though it does not scale as well to large numbers of observations 

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.swarmplot(x = 'day',y ='total_bill',data = df)
```
